# Materials for nuclear reactors

```python
import openmc
import openmc.data
import matplotlib.pyplot as plt
openmc.config['cross_sections'] = "/home/faustom/nlibs/endfb-viii.0-hdf5/cross_sections.xml"
```

## Water

```python
water = openmc.Material(name="water")
water.add_element("O", 1)
water.add_element("H", 2)
water.set_density("g/cc", 1.0)

fig, ax0 = plt.subplots()
openmc.plot_xs("H1", ["scatter", "capture"], axis=ax0)
openmc.plot_xs("O16", ["scatter", "capture"], axis=ax0)
plt.legend(["H1 scatter", "H1 capture", "O16 scatter", "O16 capture"])
plt.grid()
plt.title("Cross Sections for H1 and O16")
plt.draw()
openmc.plot_xs(water, ["scatter", "capture"])
plt.grid()
plt.xlim(0.01, 1.e6)
plt.ylim(1.e-5, 10)
plt.draw()
```

## Boron (Z=5)

Natural boron is made up of 19.65 % of B10 and 80.35 % of B11

```python
boron = openmc.Material(name="B4C")
boron.add_element("B", 4)
boron.add_element("C", 1)
boron.set_density("g/cc", 2.5)

fig, ax0 = plt.subplots()
openmc.plot_xs("B10", ["scatter", "absorption"], axis=ax0)
openmc.plot_xs("B11", ["scatter", "absorption"], axis=ax0)
plt.legend(["B10 scatter", "B10 absorpt", "B11 scatter", "B11 absorpt"])
plt.grid()
plt.title("Cross Sections for B10 and B11")
plt.draw()
openmc.plot_xs(boron, ["scatter", "absorption"])
plt.grid()
plt.xlim(0.01, 1.e6)
#plt.ylim(1.e-5, 10)
plt.draw()
```

## Zirconium (Z=40)

Fuel clad is usually made of some Zirconium alloy. 
Natural Zirconium has the following isotopic composition: 
- Zr90: 51.5 % 
- Zr91: 11.2 %
- Zr92: 17.1 %
- Zr94: 17.4 %
- Zr96: 2.8 %

```python
fig, ax0 = plt.subplots()
openmc.plot_xs("Zr90", ["absorption"], axis=ax0)
openmc.plot_xs("Zr91", ["absorption"], axis=ax0)
openmc.plot_xs("Zr92", ["absorption"], axis=ax0)
openmc.plot_xs("Zr94", ["absorption"], axis=ax0)
openmc.plot_xs("Zr96", ["absorption"], axis=ax0)
plt.title("Absorption XS for Zirconium isotopes")
plt.legend(["Zr90", "Zr91", "Zr92", "Zr94", "Zr96"])
plt.grid()
plt.xlim(0.01, 1.e6)
plt.draw()
```

```python
clad = openmc.Material(name="Zyrc")
clad.add_element("Zr", 1)
clad.set_density("g/cc", 6)
openmc.plot_xs(clad, ["scatter", "absorption"])
plt.grid()
plt.xlim(0.01, 1.e6)
plt.draw()
```

## UO2

Natural uranium is made up of 99.3 % of U238, 0.72 % of U235, and 0.005% of U234.  

```python
fig, ax0 = plt.subplots()
openmc.plot_xs("U238", ["capture"], axis=ax0)
openmc.plot_xs("U235", ["capture"], axis=ax0)
openmc.plot_xs("U234", ["capture"], axis=ax0)
openmc.plot_xs("U236", ["capture"], axis=ax0)
plt.title("Capture XS for Uranium isotopes")
plt.legend(["U238", "U235", "U234", "U236"])
plt.grid()
plt.xlim(0.01, 1.e6)
plt.draw()
```

```python

fig, ax0 = plt.subplots()
openmc.plot_xs("U238", ["fission"], axis=ax0)
openmc.plot_xs("U235", ["fission"], axis=ax0)
openmc.plot_xs("U234", ["fission"], axis=ax0)
openmc.plot_xs("U236", ["fission"], axis=ax0)
plt.title("Fission XS for Uranium isotopes")
plt.legend(["U238", "U235", "U234", "U236"])
plt.grid()
plt.xlim(0.01, 1.e7)
plt.draw()
```

```python
fig, ax0 = plt.subplots()
openmc.plot_xs("U238", ["nu-fission"], divisor_types=["absorption"], axis=ax0)
openmc.plot_xs("U235", ["nu-fission"], divisor_types=["absorption"], axis=ax0)
openmc.plot_xs("U234", ["nu-fission"], divisor_types=["absorption"], axis=ax0)
openmc.plot_xs("U236", ["nu-fission"], divisor_types=["absorption"], axis=ax0)
plt.title("nu-fission / absorption ratio for Uranium isotopes")
plt.legend(["U238", "U235", "U234", "U236"])
plt.grid()
#plt.xlim(0.01, 1.e7)
plt.draw()
```

```python
fuel = openmc.Material(name="UO2(5%")
fuel.add_element("U", 1, percent_type="ao", enrichment=5.)
fuel.add_element("O", 2)
fuel.set_density("g/cc", 10.4)

openmc.plot_xs(fuel, ["scatter", "capture", "fission"])
plt.grid()
plt.xlim(0.01, 1.e6)
plt.draw()
```
